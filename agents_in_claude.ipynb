```python
from langchain.llms import Anthropic
from langchain.agents import load_tools, initialize_agent
from langchain.tools import AIPluginTool
```

```python
PREFIX = """\n\nHuman: Answer the following questions as best you can. You have access to the following tools:"""
SUFFIX = """Begin!
Question: {input}
\n\nAssistant:
Thought:{agent_scratchpad}"""
```

```python
tool = AIPluginTool.from_plugin_url("https://www.klarna.com/.well-known/ai-plugin.json")
```

```python
llm = Anthropic(temperature=0)
tools = load_tools(["requests_all"] )
tools += [tool]
agent_kwargs = {"prefix": PREFIX, "suffix": SUFFIX}
agent_chain = initialize_agent(tools, llm, agent="zero-shot-react-description", verbose=True, agent_kwargs=agent_kwargs)
```

```python
agent_chain.run("I'm a woman of size M, I need a long bridesmaid dress for a summer wedding. Propose options in dark green.")
```
